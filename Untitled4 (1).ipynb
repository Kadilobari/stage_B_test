```python
# split the features and target into train and test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# Normalize the feature dataset
sc = MinMaxScaler ()
sc = sc.fit (x_train)
x_train_scaled = sc.transform (x_train)
x_test_scaled = sc.transform (x_test)
```

```python
# Train the model with linear regression
model = LinearRegression ()
model.fit (x_train_scaled, y_train)
```

```python
# evaluate MAE for the training set
linear_reg_mae = mean_absolute_error (y_train, model.predict (x_train_scaled))

print (f"Linear Regression mae:{linear_reg_mae})")
```

```python
linear_reg_mae = mean_absolute_error(y_train, model.predict(x_train_scaled))

print("Linear Regression MAE:", round(linear_reg_mae, 3))
```

```python
# Root Mean Squared Error for the Training Set
linear_reg_mse = mean_squared_error(y_train, model.predict(x_train_scaled))

print("Linear Regression MSE:", round(linear_reg_mse, 3))
```

```python
mse = linear_reg_mse
rmse = math.sqrt (mse)
print (round (rmse, 3))
```

```python
# Mean Absolute Error for the Test set
linear_reg_mae = mean_absolute_error(y_test, model.predict(x_test_scaled))

print("Linear Regression MAE:", round(linear_reg_mae, 3))
```
